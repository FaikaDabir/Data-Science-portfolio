```python
import copy

# 1. Define states (0 is the blank space)
initial_state = [1, 2, 3, 0, 4, 6, 7, 5, 8]
goal_state    = [1, 2, 3, 4, 5, 6, 7, 8, 0]

# 2. Heuristic function: Count misplaced tiles
def get_heuristic(state):
    count = 0
    for i in range(len(state)):
        if state[i] != 0 and state[i] != goal_state[i]:
            count += 1
    return count

# 3. Generate neighbors
def get_neighbors(state):
    neighbors = []
    blank_idx = state.index(0)
    row, col = divmod(blank_idx, 3)

    moves = []
    if row > 0: moves.append(-3) # Up
    if row < 2: moves.append(3)  # Down
    if col > 0: moves.append(-1) # Left
    if col < 2: moves.append(1)  # Right

    for move in moves:
        new_state = list(state)
        target_idx = blank_idx + move
        # Swap
        new_state[blank_idx], new_state[target_idx] = new_state[target_idx], new_state[blank_idx]
        neighbors.append(new_state)
    return neighbors

# 4. Hill Climbing Logic
current = initial_state
while True:
    current_h = get_heuristic(current)
    print(f"Current State: {current} | Misplaced: {current_h}")

    if current_h == 0:
        print("Goal reached!")
        break

    # Find the best neighbor
    neighbors = get_neighbors(current)
    best_neighbor = min(neighbors, key=get_heuristic)

    # If the best neighbor isn't better than current, we are stuck
    if get_heuristic(best_neighbor) >= current_h:
        print("Stuck in local optimum. No better moves.")
        break

    current = best_neighbor
```
